```python
import pandas as pd
import bar_chart_race as bcr
```

```python
#reading the dataset - covid dataset 

df = pd.read_csv("https://raw.githubusercontent.com/sujitkadam91/Animated-and-Story-Telling-plot-For-COVID-19-Data-set-Using-Python/master/corona_dat.csv")
```

```python
print(df['date'].max())
print(df['date'].min())
```

```python
df = df.set_index('date')
```

```python
df.tail()
```

```python
df.isnull().sum()
```

```python
df1 = df.cumsum(skipna=False)
```

```python
df1.tail()
```

```python
country_max = df1.max()
country_max = country_max.sort_values()
```

```python
print(country_max.tail(10))
```

```python
#selected the 8 countries with the maximum count + India and China
```

```python
df_plot = df1[['India','China','US','Spain','Italy','Germany','Russia','Brazil','France','United Kingdom']]
```

```python
bcr.bar_chart_race(
    df=df_plot, 
    title = "Total Covid Count Variation by Country", 
    sort = 'desc',
    n_bars = 10, 
    fixed_order = False, 
    fixed_max = True, 
    steps_per_period=10,
    label_bars=True,
    period_label={'x': .99, 'y': .25, 'ha': 'right', 'va': 'center'},
    period_summary_func=lambda v, r: {'x': .99, 'y': .18,
                                      's': f'Total count: {v.nlargest(6).sum():,.0f}',
                                      'ha': 'right', 'size': 8, 'family': 'Courier New'},
    #perpendicular_bar_func='median',
    cmap = 'prism'
    #cmap='dark12',
)
```
